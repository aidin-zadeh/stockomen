```python
import os, inspect, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import plotly.offline as pltly
import plotly.graph_objs as go
import plotly.tools as tls
import seaborn
import missingno
import warnings
```

```python
# get project root directory
curr_dir = os.path.dirname(inspect.getabsfile(inspect.currentframe()))
root_dir = os.path.dirname(curr_dir)
sys.path.insert(0, 0)
```

## Load data
We load the stock datasets to a pandas dataframe:

```python
# load data
fname = os.path.join(root_dir, "stockomen", "data", "int", "stock.csv")
df = pd.read_csv(fname)
df.head(5)
```

## Make copy of data
Once we loaded data, we make a copy of the data frame first. This allows us to modify data freely while we have always the original data for any further reference:

```python
df_clean = df.copy()
```

## Check on missing values:
Before digging into data features, we check on the data for missing values. The missing or null values is one of the most commong issue with real world data. The missing values commonly appear in situations when or where a data is not filled properly, the data is not available ore there is a computational error

In pandas terminology, the missing values are referred by `NaN` values and can simply look up by:

```python
df.isnull().values.any()
```

This shows that thera some missing values in the data frame. We use `missingno` module to further analyse the missing values in the data frame. `missingno`is a python package to perform missing data visualizations.

```python
# replace all NaN values to -1
df = df.replace(-1, np.NaN)
missingno.matrix(df=df, figsize=(20, 14))
df = df_clean.copy()
```

## Descriptive statistics
To explore data, firstly, we take a look at descriptive statistics of stocks data:

```python
df.describe()
```

```python
# set seaborn default parameters
seaborn.set()
_ = df.hist(sharex=False, sharey=False,figsize=(14, 12))
plt.show()
```

## Feature distribution
To further analyze data by featur distribution as we plot the histogram of each column:

```python
plt.scatter(df['Subjectivity'], df['Label'])
plt.xlabel('Subjectivity')
plt.ylabel('Stock Price Up or Down 0: Down, 1: Up')
plt.show()
plt.scatter(df['Objectivity'], df['Label'])
plt.xlabel('Objectivity')
plt.ylabel('Stock Price Up or Down 0: Down, 1: Up')
plt.show()
df['Subjectivity'].plot('hist')
plt.xlabel('Subjectivity')
plt.ylabel('Frequency')
plt.show()
df['Objectivity'].plot('hist')
plt.xlabel('Subjectivity')
plt.ylabel('Frequency')
plt.show()
print("Size of the Labels column")
print(df.groupby('Label').size())
```

## Correlation analysis

```python
colormap = plt.cm.afmhot
plt.figure(figsize=(16,12))
plt.title('Pearson correlation of continuous features', y=1.05, size=15)
seaborn.heatmap(df.corr(),linewidths=0.1,vmax=1.0, square=True, 
            cmap=colormap, linecolor='white', annot=True)
plt.show()

```

```python
%matplotlib inline

pltly.init_notebook_mode(connected=True)


bin_col = df.columns
zero_list = []
one_list = []
for col in bin_col:
    zero_count = 0
    one_count = 0
    for ix, val in df[col].iteritems():
        if df.loc[ix, 'Label'] == 0:
            zero_count += 1
        else:
            one_count += 1
    zero_list.append(zero_count)
    one_list.append(one_count)
    
trace1 = go.Bar(
    x=bin_col,
    y=zero_list ,
    name='Zero count'
)
trace2 = go.Bar(
    x=bin_col,
    y=one_list,
    name='One count'
)

data = [trace1, trace2]
layout = go.Layout(
    barmode='stack',
    title='Count of 1 and 0 in binary variables'
)

fig = go.Figure(data=data, layout=layout)
pltly.iplot(fig, filename='stacked-bar')
```
